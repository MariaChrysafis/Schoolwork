```python
import numpy as np
import matplotlib.pyplot as plt

def get_means(f):
    tot = 0
    ans = []
    for n in range(10_000):
        tot += f()  # Call f without passing arguments since f is a function returning random values
        ans.append(tot / (n + 1))
    return ans  # Return the list of means

# Plot for Normal Random Variables
plt.title("Normal Random Variables")
plt.plot(get_means(lambda: np.random.normal(0, 1)))
plt.show()
# Plot for Cauchy Random Variable
plt.title("Cauchy Random Variable")
plt.plot(get_means(lambda: np.random.standard_cauchy()))

plt.show()
```
